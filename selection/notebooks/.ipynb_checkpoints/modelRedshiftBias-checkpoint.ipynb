## Luminosity Cut



```python
def rm_char_luminosity(z):
    lnz  = np.log(z)
    res0 = 22.44+3.35*lnz+0.273*lnz**2-0.0618*lnz**3-0.0227*lnz**4
    res1 = 22.94+3.08*lnz-11.22*lnz**2-27.11*lnz**3-18.02*lnz**4
    res = np.where(z>0.5,res1,res0)
    return res
```

```python
Z1 = [ 2.85703803, 23.35451485]
Z2 = [ -0.13290108,   1.11453393,  -3.74351207,   6.96351527, -10.04310863,  12.60755443,  -7.85820367,   0.92175766]
def char_luminosity(z):
    return np.poly1d(Z1)(np.log(z)) + np.poly1d(Z2)(z)
```

### Load One Healpix

```python
import pandas as pd
def upload_dataFrame(infile,keys='members'):
    hdf = pd.HDFStore(infile, mode='r')
    df1 = hdf.get(keys)
    hdf.close()
    data = Table.from_pandas(df1)
    return data
```

```python
import smatch
zb = np.linspace(-0.1, 0.1, 21)
outdir = '/data/des61.a/data/johnny/DESY3/desy3/data/spt/cutouts/'
outfile_base = outdir+'y3_gold_2.2.1_wide_sofcol_run_join_2500d_sptecs_hpx2_{:05d}.hdf5'

def zoffset(z,z0):
    return (z-z0)/(1+z0)

def get_prob(data):
    dm = np.abs(data['sof_cm_mag_corrected_r'] - np.max(data['sof_cm_mag_corrected_r']))
    return dm/np.sum(dm)

def get_stats(gal):
    zres = dict()
    r1,r2,nobj,zcls = [], [],[],[]
    for cid in np.unique(gal['CID']):
        mask = gal['CID'] == cid
        g = gal[mask]
        if len(g)>0:
            zoff = zoffset(g['z_mean_dnf'],g['zcls'])
            prob = get_prob(g)
            r1.append(np.median(zoff))
            r2.append(np.average(zoff,weights=prob))
            zcls.append(np.mean(g['zcls']))
            nobj.append(np.count_nonzero(mask))
            
    zres['zcls'] = zcls
    zres['avg'] = r2
    zres['med'] = r1
    zres['nobj'] = nobj
    return zres

def get_bias(hpix_number):
    outfile = outfile_base.format(hpix_number)
    ci = upload_dataFrame(outfile,keys='cluster')
    gi = upload_dataFrame(outfile,keys='members')
    
    ###########
    # sky match
    nside=4096 # healpix nside
    maxmatch=-1 # return all matches

    indices_all = []
    ra,dec = gi['ra'], gi['dec']
    rac, decc = np.array(ci['RA']), np.array(ci['DEC'])

    # within 500kpc from the cluster center
    rmax = np.array(0.5*(ci['rmax']/8.)/60.)

    sky = smatch.match(rac, decc, rmax, ra, dec, nside=nside, maxmatch=maxmatch)
    
    gal = gi[sky['i2']]
    gal['CID'] = sky['i1']
    gal['ang_dist'] = 60*np.arccos(sky['cosdist'])*180/np.pi
    gal['R'] = gal['ang_dist']/(ci[sky['i1']]['rmax']/8.)
    gal['zcls'] = ci[sky['i1']]['REDSHIFT']

    cat = ci[np.unique(sky['i1'])]
    cat['CID'] = np.unique(sky['i1'])

    # apply redshift and luminosity cut
    # zcut 
    zcut = np.abs(gal['z_mean_dnf']-gal['zcls']) < 0.1*(1+gal['zcls'])

    # luminosity cut
    mcut = gal['sof_cm_mag_corrected_r'] <= rm_char_luminosity(gal['zcls'])+1.0

    gal = gal[mcut&zcut]
    print('The     galaxy size is: %.2f M'%(len(gi)*1e-6))
    print('The new galaxy size is: %i'%(len(gal)))
    
    #######################
    ## redshift bias stats
    zres = get_stats(gal)
    return zres
```

```python
tiles = [16,20,32,33,34,35,38,44,45,46]

res = []
for hpix in tiles:
    print('Tile %i'%hpix)
    res.append(get_bias(hpix))
    
    print('-------------')
```

```python
cols = ['zcls','avg','med','nobj']
out = dict().fromkeys(cols)

for col in cols:
    out[col] = np.hstack([r[col] for r in res])
    
out['bias'] = out['med']/(1+out['zcls'])
```

```python
def makeBins(variable,xedges):
    xbins = (xedges[1:]+xedges[:-1])/2
    indices = [ np.where((variable >= xedges[i]) & (variable <= xedges[i + 1]))[0] for i in range(len(xedges)-1)]
    return indices, xbins
```

```python
def remove_outliers(x):
    q25,q75 = np.nanpercentile(x,[25,75])
    iqr     = q75-q25
    lo, up  = q25-1.5*iqr, q75+1.5*iqr
    mask    = (x<up)&(x>lo)
    return mask

def fit_gauss(z):
    mask = remove_outliers(z)
    zfit = z[mask]
    zmin = np.nanpercentile(zfit,16)
    zmax = np.nanpercentile(zfit,84)
    res  = np.array([np.nanmean(zfit),np.nanstd(zfit),zmin,zmax,1-1.*np.count_nonzero(mask)/len(z)])
    return res
```

```python
out2 = dict()
for col in cols:
    out2[col] = out[col][out['nobj']>10.]
```

```python
zmin, zmax, dx = 0.05, 1.3, 0.10

zbins   = np.arange(zmin,zmax+dx,dx)
keys, zb  = makeBins(out['zcls'],zbins)

## Computing model variables
res  = np.array([fit_gauss(out['bias'][idx]) for i,idx in enumerate(keys)])
bias = res[:,0]
sigma= res[:,1]
zlow = res[:,2]
zhigh= res[:,3]
of   = res[:,4]
```
